```python
def category(price):
    if price <= 100:
        return 'economy'
    elif price >= 200:
        return 'premium'
    else:
        return 'comfort'

sep_by_price = Rooms.copy()
sep_by_price['category'] = sep_by_price['price'].apply(category)
sep_by_price = sep_by_price.groupby('category').size().reset_index(name='count')
sep_by_price
```

```python
comfortable_rooms_ordered = Rooms.merge(Reservations,how='left', left_on='id', right_on='room_id')
comfortable_rooms_ordered['days'] = (comfortable_rooms_ordered['end_date'] - comfortable_rooms_ordered['start_date']).dt.days.fillna(0)
comfortable_rooms_ordered = comfortable_rooms_ordered[(comfortable_rooms_ordered['has_air_con']==1)&(comfortable_rooms_ordered['has_internet']==1)&(comfortable_rooms_ordered['has_kitchen']==1)&(comfortable_rooms_ordered['has_tv']==1)].groupby(['home_type','address']).agg(days=('days', 'sum'), total_fee=('total', 'sum')).reset_index()
comfortable_rooms_ordered
```

```python
avg_rooms_rating = Reservations.merge(Reviews, left_on='id', right_on='reservation_id').merge(Rooms, left_on='room_id', right_on='id')[['room_id', 'rating']].groupby('room_id').agg(avg_rating = ('rating', 'mean')).reset_index()
avg_rooms_rating
```

```python
amount_reservs_per_month = Reservations.copy()
amount_reservs_per_month['year'] = amount_reservs_per_month['start_date'].dt.year
amount_reservs_per_month['month'] = amount_reservs_per_month['start_date'].dt.month
amount_reservs_per_month = amount_reservs_per_month.groupby(['year', 'month']).size().reset_index(name='amount')
amount_reservs_per_month
```
