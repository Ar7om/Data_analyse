```python
import pandas as pd
```

```python
df = pd.read_csv('НТА Тест задание 1.csv', sep=';')
df.head()
```

```python
df['rang'] = df.groupby('dep')['salary'].rank(method='dense', ascending=False)
df.sort_values(by=['dep', 'salary'], ascending=[True, False]).reset_index(drop=True)
```

```python
result = df[df['rang'] == 1]['emplid'].reset_index(drop=True)
result.head()
```
