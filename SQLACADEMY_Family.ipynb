```python
good_type_2005 = GoodTypes.merge(Goods, how='left', left_on='good_type_id', right_on='type').merge(Payments[Payments['date'].dt.year == 2005], how='left', left_on='good_id', right_on='good')
good_type_2005 = good_type_2005.groupby('good_type_name').apply(
    lambda x: (x['amount']*x['unit_price']).sum()
).reset_index(name='costs').sort_values('costs', ascending=False).reset_index(drop=True)
good_type_2005
```

```python
not_payed_good_type_in_2005 = GoodTypes.merge(Goods, how='left', left_on='good_type_id', right_on='type').merge(Payments[Payments['date'].dt.year == 2005], how='left', left_on='good_id', right_on='good')
not_payed_good_type_in_2005 = not_payed_good_type_in_2005.groupby('good_type_name').filter(lambda x: x['payment_id'].isnull().all())[['good_type_name']].drop_duplicates().reset_index(drop=True)
not_payed_good_type_in_2005
```

```python
costs_in_june_2005 = Payments[(Payments['date'].dt.year == 2005)&(Payments['date'].dt.month == 6)].merge(FamilyMembers, how='right', left_on='family_member', right_on='member_id')
costs_in_june_2005['costs'] = costs_in_june_2005['amount']*costs_in_june_2005['unit_price']
costs_in_june_2005 = costs_in_june_2005.groupby('member_name')['costs'].sum()
costs_in_june_2005 = pd.DataFrame(costs_in_june_2005)
costs_in_june_2005.sort_values(by='costs', ascending=False).reset_index()
```

```python
most_expensive_delicacies = Payments.merge(Goods, left_on='good', right_on='good_id').merge(GoodTypes, left_on='type', right_on='good_type_id')
most_expensive_delicacies = most_expensive_delicacies[most_expensive_delicacies['good_type_name']=='delicacies']
most_expensive_delicacies = most_expensive_delicacies[most_expensive_delicacies['unit_price']==(most_expensive_delicacies['unit_price'].max())][['good_name', 'unit_price']]
most_expensive_delicacies
```

```python
count_goods_more_one = Payments.merge(Goods, left_on='good', right_on='good_id')[['good_name']]
count_goods_more_one = count_goods_more_one.groupby('good_name').size().reset_index(name='count')
count_goods_more_one = pd.DataFrame(count_goods_more_one)[count_goods_more_one['count'] > 1]
count_goods_more_one.sort_values(by='count', ascending=False)
```

```python
costs_for_entertainment = Payments.merge(FamilyMembers, left_on='family_member', right_on='member_id').merge(Goods, left_on='good', right_on='good_id').merge(GoodTypes, left_on='type', right_on='good_type_id')
costs_for_entertainment = costs_for_entertainment[costs_for_entertainment['good_type_name'] == 'entertainment'][['status', 'member_name', 'amount', 'unit_price']]
costs_for_entertainment = costs_for_entertainment.groupby(['status', 'member_name']).apply(lambda x: (x['amount']*x['unit_price']).sum()).reset_index(name='costs')
costs_for_entertainment.sort_values(by='costs', ascending=False)
```
