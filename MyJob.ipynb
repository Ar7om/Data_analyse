```python
import pandas as pd
import os
```

```python
file_path = r'C:\Users\F\Downloads\Telegram Desktop\АоРПИ 61.13-АПС.СОУЭ.xlsx'
```

```python
rows_need = [15, 23]
```

```python
excel_file = pd.ExcelFile(file_path)
sheets_names = excel_file.sheet_names
```

```python
sheets_names[:5]
```

```python
len(sheets_names)
```

```python
need_data = []

for vkladka in sheets_names:
    df = pd.read_excel(file_path, sheet_name=vkladka, header=None)

    for q in rows_need:
        need_value = df.iloc[q-1, 0]
        need_data.append([vkladka, q, need_value])

result_df = pd.DataFrame(need_data, columns=['Вкладка', 'Строка', 'Значение ячейки'])
```

```python
result_df.head()
```

```python
result_df.to_excel('Итог_по_АоРПИ.xlsx', index=False)
```

```python
df = result_df
df = df.rename(columns={'Вкладка':'Head', 'Строка':'Row', 'Значение ячейки':'Value'})
df.head()
```

```python
df['Value'] = df['Value'].str.replace('.,', ';', regex=False).replace()

df = (
     df.assign(Value=df['Value'].str.split(';'))  # разделяем на списки
    .explode('Value')  # создаем новые строки для каждого элемента списка
    .reset_index(drop=True)  # сбрасываем индекс
)
```

```python
df.head()
```

```python
df.to_excel('После разделения и на кол-во.xlsx', index=False )
```

```python
result_data = []

# Обрабатываем каждую строку исходных данных
for _, row in df.iterrows():
    if row['Row'] == 15:
        # Для строк с Row=15 добавляем как есть
        result_data.append({
            'Head': row['Head'],
            'Row': row['Row'],
            'Value': row['Value'],
            'Pasport': ''
        })
    elif row['Row'] == 23:
        # Для строк с Row=23 добавляем с пустым Value
        result_data.append({
            'Head': row['Head'],
            'Row': row['Row'],
            'Value': '',
            'Pasport': row['Value']
        })

# Создаем финальный DataFrame
result = pd.DataFrame(result_data)
```
