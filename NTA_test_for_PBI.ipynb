```python
import pandas as pd
```

```python
df = pd.read_excel('НТА Тест PBI Пассажиропоток.xlsx')
df.head()
```

```python
types = df['Тип ТК'].unique()
types
```

```python
need_list = []
for i in types:
    if 'БК' in i:
        category = 'БК'
        commission = 0.075
    elif 'НАЛ' in i:
        category = 'НАЛ'
        commission = 0.02
    elif 'ЭК' in i:
        category = 'ЭК' 
        commission = 0.07
    elif 'ПБ' in i:
        category = 'ПБ'
        commission = 0.06
    else:
        category = 'Другое'
        commission = 0
    
    need_list.append({'Тип ТК': i, 'Категория': category, 'Комиссия': commission})
need_list
```

```python
result_df = pd.DataFrame(need_list)
result_df
```

```python
result_df.to_excel('Типы ТК.xlsx', index=False)
```
