```python
import pandas as pd
from datetime import datetime
```

```python
df = pd.read_csv('new.csv', sep='\t')
df.head()
```

```python
df.columns
```

```python
df = df.drop(['Unnamed: 0'], axis=1)
```

```python
df.head()
```

```python
df.info()
```

```python
df['date_open'] = pd.to_datetime(df['date_open'], format='%Y-%m-%dT%H:%M:%S.%fZ')
df.head()
```

```python
df['date_close'] = pd.to_datetime(df['date_close'], format='%Y-%m-%dT%H:%M:%S.%fZ', errors='coerce')
df.head()
```

```python
df.info()
```

```python
now_date = datetime.now()
working_w = df[
    (df['date_open']<=now_date) & (df['date_close'].isna()|(df['date_close']>now_date))
]
```

```python
working_w.head()
```

```python
result = working_w.groupby('city').size().reset_index(name='count')
result
```

```python
result = result[result['count']>=80].sort_values(by='count', ascending=False)
result
```
